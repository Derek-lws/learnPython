```python
from matplotlib import pyplot as plt
import math
t = np.linspace(0, math.pi, 1000)
x = np.sin(t)
y = np.cos(t) + np.power(x, 2.0/3)
plt.plot(x, y, color='pink', linewidth=2, label='h')
plt.plot(-x, y, color='pink', linewidth=2, label='-h')
plt.xlabel('t')
plt.ylabel('h')
plt.ylim(-2, 2)
plt.xlim(-2, 2)
plt.legend()#图例
plt.show()
```
