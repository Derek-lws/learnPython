策略：
选取股代码为601009的股票，每个自然月的第一个交易日买入100股
每年最后一个交易日卖出所有股票且最后一个交易日卖出所有股票
求07年03月至今收益是多少

```python
price_last = df['open'][-1]#最后一个交易日股票价格
df = df['2007-07':'2020-03']#时间筛选
df_monthly = df.resample('M').first()#按月第一个交易日
#重新采样(pandas对象都会有这个方法),
#在给定的时间单位内重取样,
#是一个对常规时间序列数据重新采样和频率转换的便捷的方法
df_yearly = df.resample('A').last()[:-1]#除最后一年每年最后一个交易日
```

```python
df_monthly
```

```python
cost_money = 0#购买股票的资金
hold = 0#持有股票数量
for year in range(2007,2020):
    cost_money += df_monthly[str(year)]['open'].sum()*100#买入
    hold += len(df_monthly[str(year)]['open'])*100#计算持有股票数
    if year !=2019:
        cost_money -=df_yearly[str(year)]['open'][0] * hold
        hold = 0
    print(cost_money)
cost_money -= hold * price_last#全部卖出
print(-cost_money)
```
